### Data Cleaning and Feature Engineering

```python
customer.isna().sum()
```

Few observation about data
- So we have data set with 2240 rows and 29 columm.
- There are 24 missing values for column `Income`.
- There are two categorical columns `Education` and `Marital_Status`.
- `Dt_Customer` is the date when a customer was enrolled with company.



Now, we don't have any missing values in dataset. But number of rows or observaton is reduced.

Since `Dt_Customer` is not in date format. I need to convert it into date format.

```python
customer["Dt_Customer"] = pd.to_datetime(customer["Dt_Customer"])
```

Let's look at summary of numerical columns in dataset:

```python
customer.describe().T
```

Frequency of each Category of categoical column:


```python
print("Education Categories Frequency")
print( customer["Education"].value_counts())
print("\n")
print("Marital Status Categories Frequency")
print(customer["Marital_Status"].value_counts())
```

```python
# Creating new variable days_joined assuming last join date is today (day of all these analysis)
customer["last_date"] = max(customer["Dt_Customer"])
customer["days_joined"] = (customer["last_date"] - customer["Dt_Customer"]).dt.days

# Creating new variable Age as of 2014
customer["Age"] = 2014 - customer["Year_Birth"]

# Creating new column Spent as sum of all money spent in 6 types of products in last 2 years
customer["Spent"] = customer["MntWines"]+ customer["MntFruits"]+ customer["MntMeatProducts"]+ customer["MntFishProducts"]+ customer["MntSweetProducts"]+ customer["MntGoldProds"]

# Creating new column AcceptedCampaign as no. of campaigns in which customer accepted offer
customer["AcceptedCampaign"] = customer["AcceptedCmp1"]+ customer["AcceptedCmp2"] + customer["AcceptedCmp3"] + customer["AcceptedCmp4"]+ customer["AcceptedCmp5"]+ customer["Response"]


# 
customer.head()
```
